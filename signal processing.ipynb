```python
import tensorflow as tf
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
X = np.random.random((1000,1))*6-3
```

```python
Y = (np.cos(2*X) + np.sin(4*X)).flatten() 
Y = Y.reshape((1000,))
```

```python
plt.scatter(X,Y)
```

```python
model = tf.keras.Sequential([
                             tf.keras.layers.Dense(64, input_shape = (1,), activation = 'relu'),
                             tf.keras.layers.Dense(128, input_shape= (64,),activation= 'relu'),
                             tf.keras.layers.Dense(1)
])

opt = tf.keras.optimizers.Adam(0.01)
model.compile(optimizer = opt, loss = 'mse')

r = model.fit(X, Y, epochs = 100)
```

```python
plt.plot(r.history['loss'])
```

```python
k = np.linspace(-3,3, 100)
k = k.reshape(-1,1)
yhat = model.predict(k).flatten()
yhat.shape
```
